# Nature type from the Norwegian Environmental Agency

In this notebook, we work on a dataset with Nature Types from NEA and Nature in Norway Type for the whole Norway.

### 1) Nature Types 

They are determined according to the Norwegian Environment Agency's instructions.

According to NEA, Nature Types are an uniform type of nature prioritized for mapping. 

A field survey of nature types on land follow strict instructions.

The instructions describe the mapping of 111 nature types, of which 83 are red-listed in accordance with the Norwegian Red List for Nature Types 2018 (Artsdatabanken), while 28 have been determined based on the recommendation of an expert group consisting of the Norwegian Institute for Natural Research (Nina), the Norwegian Institute for Bioeconomy (Nibio) and NTNU The Science Museum.

The different Nature Types are described here: 
https://www.miljodirektoratet.no/ansvarsomrader/overvaking-arealplanlegging/naturkartlegging/naturtyper/

For visualisation, please, follow this link: https://miljoatlas.miljodirektoratet.no

For downloading, please, follow this link: https://karteksport.miljodirektoratet.no/

More info: https://www.miljodirektoratet.no/naturbase

### 2) Nature in Norway Type (NiN).

NiN Type provide a description about the locality's condition and natural diversity

More information: https://artsdatabanken.no/NiN/Natursystem/Typeinndeling

### 3) Purpose of this notebook

- Read the geojson, 

- convert it CRS into `epsg 3035`, 

- encoded the nature type into integer.

- add the boundary box of each for each its polygons.

This processed geojson will be used to crop and combine large xarray with polygons.

# Geographical projection transformation

From lon/lat to epsg 3035, https://land.copernicus.eu/imagery-in-situ/eu-dem/eu-dem-v1.1?tab=metadata

```python
df = df.to_crs(epsg=3035)
```

# Joining an encoded label of nature type to the original dataset

```python
labels = df.naturtypeKode.drop_duplicates().values
```

```python
vocab = sorted(set(labels)) # unique and sorted
```

```python
vocab_size = len(vocab)
```

```python
# map words to unique indices
word2idx = {key: ind for ind, key in enumerate(vocab)} 
```

```python
# REM: I should get 111 Nature Type + some missing values or unknown type, but got 176...
len(word2idx)
```

```python
word2idx
```

Remarks: nature type `ntyp_I00` described in our encoding as `175` represent missing values.

```python
labels = df.naturtypeKode.values
```

```python
encoded_labels = [word2idx[key] for key in labels]
```

```python
df["nature_encoded"] = pd.DataFrame(encoded_labels)
```

```python
df.head()
```

# Joining a boundary box  to the original dataset

This will be used when clipping the xarray dataset with all the features.

Reason:

- Clipping large xarray w/ polygon is not possible due to the need of storing large amount of data into memory.

- However, xarray enables filtering coordinates with boundary box in a efficient way.

- So, to clip on an array, we will first crop on the boundary box, and then clip on the polygon.

ref: 

https://corteva.github.io/rioxarray/stable/examples/clip_geom.html

https://gis.stackexchange.com/questions/328128/extracting-data-within-geometry-shape

https://github.com/pydata/xarray/issues/501#issuecomment-126461466

### Calling dask

```python
ddf = dask_geopandas.from_geopandas(df, npartitions=10)
```

### Boundary box

```python
bbox = ddf.geometry.bounds.compute()
```

```python
ddf2 = pd.concat([ddf.compute(),bbox],axis=1)
```

```python
ddf2
```

```python
# This path is for frozen data 
#res_path = '/storage/nilu/DIGITAL/Project data/ETC-DI/321_habitat_mapping/Miljødirektorat/NT_no_3035_geojson'
res_path = '/your/path/'

if not os.path.exists(res_path):
    os.makedirs(res_path)
    
ddf2.to_file(os.path.join(res_path,'NT_no_3035_geojson.json'), driver='GeoJSON')
```
